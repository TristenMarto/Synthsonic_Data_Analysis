```python
home_folder = Path.cwd().parents[1]
path = home_folder.joinpath(home_folder, 'CSV_results')
df_info = pd.read_csv(path.joinpath('dataset_info.csv'))
df = pd.read_csv(path.joinpath('Cross_validation', 'Cross_validation.csv'), index_col=0)
HR = HandleResults()
```

```python
def ranking_table(df) :
    
    metrics = ['balanced_accuracy','G_mean','f1','precision','recall','pr_auc']
    df2 = pd.concat([HandleResults().final_ranking(df,metric) for metric in metrics], axis=1).rename(columns={'index': 'Oversampler'})
    df3 = pd.concat([df2.iloc[:,[0,i]].sort_values(by=df2.columns[i], ascending=False).reset_index(drop=True) for i in np.arange(1,13,2)], axis=1)
```

```python
metrics = ['balanced_accuracy','G_mean','f1','precision','recall','pr_auc']
df2 = pd.concat([HR.final_ranking(df,metric) for metric in metrics],axis=1).reset_index().rename(columns={'index': 'Oversampler'})
df3 = pd.concat([df2.iloc[:,[0,i]].sort_values(by=df2.columns[i], ascending=False).reset_index(drop=True) for i in np.arange(1,13,2)], axis=1)
df3.index = np.arange(1, len(df3)+1)
df3
# print(df3.to_latex())
```
