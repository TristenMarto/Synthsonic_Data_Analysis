```python
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import RandomOverSampler, SMOTE, SMOTENC, SVMSMOTE, ADASYN, BorderlineSMOTE
from sklearn.datasets import make_classification

import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parents[1]))
from handle_data import HandleData
```

```python
HR = HandleData()
X,y = make_classification(weights=[0.9,0.1],n_classes=2, n_features=20, n_samples=1000, class_sep=1,random_state=5)
HandleData().pca_plot(X,y)
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE, SMOTENC, SVMSMOTE, ADASYN, BorderlineSMOTE
o = SMOTE(sampling_strategy=1, random_state=5)
X_g, y_g = o.fit_resample(X,y)

HR.pca_plot(X_g, y_g)
```

```python
o = BorderlineSMOTE(sampling_strategy=1, random_state=5)
X_g, y_g = o.fit_resample(X,y)

HR.pca_plot(X_g, y_g)
```

```python
o = ADASYN(sampling_strategy=1, random_state=5)
X_g, y_g = o.fit_resample(X,y)

HR.pca_plot(X_g, y_g)
```
