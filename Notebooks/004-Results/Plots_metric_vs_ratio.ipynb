```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import sys
sys.path.append(str(Path.cwd().parents[1]))
from handle_results import HandleResults
```

```python
home_folder = Path.cwd().parents[1]
plot_folder = home_folder.joinpath(home_folder, 'Plots')
path = home_folder.joinpath(home_folder, 'CSV_results')
df_info = pd.read_csv(path.joinpath('dataset_info.csv'))
df = pd.read_csv(path.joinpath('Cross_validation', 'Cross_validation.csv'), index_col=0)
HR = HandleResults()
```

# dataset

```python
a = list(df_info.sort_values('imbalance_ratio').loc[:,'dataset'])
b = HR.select_best(df, 'f1')
b['dataset'] = pd.Categorical(b['dataset'], a)
b = b.sort_values('dataset')
```

```python
fig, ax = plt.subplots(figsize=(12,7))

ax.plot(b.dataset.unique(), b.loc[b.oversampler == 'synthsonic','G_mean'])
ax.plot(b.dataset.unique(), b.loc[b.oversampler == 'SMOTE','G_mean'])
ax.plot(b.dataset.unique(), b.loc[b.oversampler == 'NoOversampling','G_mean'])
```
