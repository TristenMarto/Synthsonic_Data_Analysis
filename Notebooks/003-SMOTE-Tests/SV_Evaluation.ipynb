```python
class synthsonic(sv.OverSampling) :
    
    def __init__(self,param1=5,param2=5, random_state=None) :
        super().__init__()
        
        self.param1 = param1
        self.param2 = param2
        self.random_state = random_state
        
    @classmethod
    def parameter_combinations(cls) :
        return [{'param1': 1, 'param2': 'a'},
                {'param1': 2, 'param2': 'b'},
                {'param1': 3, 'param2': 'c'}]
    
    
    def find_minority(self, X, y):

        unique, counts = np.unique(y, return_counts=True)
        self.class_stats = dict(zip(unique, counts))
        
        self.min_label = unique[0] if counts[0] < counts[1] else unique[1]
        self.maj_label = unique[1] if counts[0] < counts[1] else unique[0]
        
    
    def sample(self, X, y) :
        
        # Find minority class
        self.find_minority(X,y)
        
        self.X_min = X[y == self.min_label]
        
        # determine n_samples
        self.n_samples = self.class_stats[self.maj_label] - self.class_stats[self.min_label]
        
        # fit model
        kde = KDECopulaNNPdf()
        kde = kde.fit(self.X_min)
        
        
        # sample
        x1, weights = kde.sample(n_samples=self.n_samples)
        
        X_samp = np.vstack([X,x1])
        y_samp = np.hstack([y, [self.min_label]*self.n_samples])
        
        return X_samp, y_samp

    def get_params(self) :
        return {'param1': self.param1, 'param2': self.param2}
    
    
    
def plotter(X,y,title) :
    
    counter = Counter(y)
    
    for label, _ in counter.items() :
        rowix = np.where(y == label)[0]
        plt.scatter(X[rowix, 0], X[rowix, 1], label=str(label))
    plt.legend()
    plt.title(title)
    plt.show()

    print(counter)
```

```python
over = synthsonic()
```

```python
dataset= imbd.load_iris0()
```

```python
X_samp, y_samp= over.sample(dataset['data'], dataset['target'])
```

```python
plotter(dataset['data'], dataset['target'], "original data")

plotter(X_samp, y_samp, "Synthsonic sample")
```
