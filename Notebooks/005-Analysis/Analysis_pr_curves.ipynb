# Kneighbors check

```python
def PR_curves(dataset, clf, over) :
    
    X,y,title = load_data(dataset)

    x = StandardScaler().fit_transform(X)

    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2'])


    counter = Counter(y)
    print(counter)


    # make plots
    f, axes = plt.subplots(2, 2, figsize=(12, 12))

    f.suptitle(f"PCA plot and PR Curve for standard and oversampled {clf}")

    axes[0,0].set_xlabel('PC1')
    axes[0,0].set_ylabel('PC2')


    for label, _ in counter.items() :
        rowix = np.where(y == label)[0]
        axes[0,0].scatter(principalComponents[rowix, 0], principalComponents[rowix, 1], label=str(label))

    axes[0,0].legend(fontsize='small')


    # do kfold cross val and plot the PR curves
    k_fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=5)

    y_real = []
    y_proba = []
    for i, (train_index, test_index) in enumerate(k_fold.split(X,y), 1):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]    

        clf.fit(X_train, y_train)
        probas = clf.predict_proba(X_test)
        precision, recall, _ = precision_recall_curve(y_test, probas[:,1])
        lab = f"Fold {i} AUC = {auc(recall, precision)}"
        axes[0,1].step(recall, precision, label=lab)
        y_real.append(y_test)
        y_proba.append(probas[:,1])

    y_real = np.concatenate(y_real)
    y_proba = np.concatenate(y_proba)
    precision, recall, _ = precision_recall_curve(y_real, y_proba)
    lab = f"Overall AUC = {auc(recall, precision)}"
    axes[0,1].step(recall, precision, label=lab, lw=2, color='black')
    axes[0,1].set_xlabel('Recall')
    axes[0,1].set_ylabel('Precision')
    axes[0,1].legend(fontsize='small')



   
    axes[1,0].set_xlabel('PC1')
    axes[1,0].set_ylabel('PC2')

    
    try :
        X_g, _ = over.sample(X,y)
    
    except :
        X_g, _ = over.fit_resample(X,y)
    
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(X_g)
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2'])

    for label, _ in counter.items() :
        rowix = np.where(y == label)[0]
        axes[1,0].scatter(principalComponents[rowix, 0], principalComponents[rowix, 1], label=str(label))

    axes[1,0].legend(fontsize='small')

    # do kfold cross val and plot the PR curves
    y_real = []
    y_proba = []
    for i, (train_index, test_index) in enumerate(k_fold.split(X,y), 1):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]    

        # generate samples
        try :
            X_train, y_train = over.sample(X_train, y_train)
        
        except :
            X_train, y_train = over.fit_resample(X_train, y_train)
        
        clf.fit(X_train, y_train)
        probas = clf.predict_proba(X_test)
        precision, recall, _ = precision_recall_curve(y_test, probas[:,1])
        lab = f"Fold {i} AUC = {auc(recall, precision)}"
        axes[1,1].step(recall, precision, label=lab)
        y_real.append(y_test)
        y_proba.append(probas[:,1])

    y_real = np.concatenate(y_real)
    y_proba = np.concatenate(y_proba)
    precision, recall, _ = precision_recall_curve(y_real, y_proba)
    lab = f"Overall AUC = {auc(recall, precision)}"
    axes[1,1].step(recall, precision, label=lab, lw=2, color='black')
    axes[1,1].set_xlabel('Recall')
    axes[1,1].set_ylabel('Precision')
    axes[1,1].legend(fontsize='small')


    f.tight_layout()
```

# Manual calculation

```python
def get_pr_curve(y_true, probas, thresholds) :
    
    xax = []
    yax = []
    for thr in thresholds :
        
        preds = Binarizer(threshold=thr).fit_transform([probas]).ravel()
        tn, fp, fn, tp = confusion_matrix(y_true, preds).ravel()
        

        precision = tp/(fp+tp)
        recall = tp/(tp+fn)

        
        xax.append(recall)
        yax.append(precision)
        
    return xax, yax
```

```python
def classify(dataset, clf) :
    X,y,title = load_data(dataset)
    over = SMOTE(random_state=5)

    # single split
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, stratify = y, random_state = 5)

    X_g, y_g = over.fit_resample(X_train, y_train)

    clf.fit(X_g, y_g)

    pred = clf.predict(X_test)
    probas = clf.predict_proba(X_test)

    print("class imbalance:",Counter(y))


    return pred, probas
```

```python
clf = KNeighborsClassifier()

pred, probas = classify('abalone', clf)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, stratify = y, random_state = 5)
print("confusion matrix\n",confusion_matrix(y_test, pred))

tp, fn, fp, tn = confusion_matrix(y_test, pred).ravel()

precision = tp/(tp+fp)
recall = tp/(tp+fn)

print("precision:",precision)
print("recall:",recall)

plt.title("Distribution of probabilities: target class")
plt.hist(probas[:,1], bins=100, range=(0,1), alpha=0.5, density=True)

thresholds = np.linspace(0,1,num=100)

xax, yax = get_pr_curve(y_test, probas[:,1], thresholds)

plt.figure(figsize=(12,7))
plt.title("PR curve Abalone")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.scatter(xax, yax)
```
